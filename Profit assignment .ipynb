```python
rsq_RD = smf.ols('RD~Administration+MarketingSpend', data = df1).fit().rsquared
vif_RD = 1/(1-rsq_RD)

rsq_AD = smf.ols('Administration~RD+MarketingSpend', data =  df1).fit().rsquared
vif_AD = 1/(1-rsq_AD)

rsq_MS = smf.ols('MarketingSpend~RD+Administration', data = df1).fit().rsquared
vif_MS = 1/(1-rsq_MS)

d1={'Varibles':['RD','AD','MS'], 'VIF':[vif_RD,vif_AD,vif_MS]}
Vif_frame = pd.DataFrame(d1)
Vif_frame
```

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q')
plt.title('normal Q-Q plot of residuals')
plt.show()
```

```python
def get_standardized_values(vals):
    return(vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))
plt.title('Residual Plot')
plt.xlabel('Standardized Fitted Vlaues')
plt.ylabel('Standardized Residual Values')
plt.show()
```

```python
model_influence = model.get_influence()
(c,_) = model_influence.cooks_distance
c
```

```python
# Plot the influencers using the stem plot
fig = plt.subplots(figsize=(20,7))
plt.stem(np.arange(len(df1)), np.round(c,3))
plt.xlabel('Row Index')
plt.ylabel("Cook's Distance")
plt.show()
```

```python
# Index and value of influencer where C>0.5
(np.argmax(c),np.max(c))
```

```python
# 2. Leverage Value using High Influence Points : Points beyond Leverage_cutoff value are influencers
from statsmodels.graphics.regressionplots import influence_plot
influence_plot(model)
plt.show()
```

```python
# Leverage Cuttoff Value = 3*(k+1)/n ; k = no.of features/columns & n = no. of datapoints
k = df1.shape[1]
n = df1.shape[0]
leverage_cutoff = 3*((k+1)/n)
leverage_cutoff
```

```python
#49 is the infulencer
df1[df1.index.isin([49])]
```

```python
# Discard the data points which are influencers and reassign the row number (reset_index(drop=True))
df2 = df1.drop(df1.index[[49]],axis = 0).reset_index(drop= True)
df2.head()
```

```python
finalmodel = smf.ols('Profit~RD+Administration+MarketingSpend',data=df2).fit()
finalmodel.summary()
```
